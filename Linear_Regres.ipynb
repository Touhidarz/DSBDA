```python
def remove_outlier(data, column):
    mean = data[column].mean()
    std = data[column].std()
    threshold = 3*std

    zScore = (data[column]-mean)/std

    no_outlier = data[np.abs(zScore) >= threshold]

    return no_outlier

```

```python
data_no_outlier = remove_outlier(data, 'rm')
data = data_no_outlier

sns.boxplot(data=data, x='rm')
```

```python
data_no_outlier = remove_outlier(data, 'medv')
data = data_no_outlier

sns.boxplot(data=data, x='medv')
```

```python
data.isnull().sum()
```

```python
data['rm'].fillna(data['rm'].mean(), inplace=True)
data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
X = data['rm'].values.reshape(-1,1)
Y = data['medv']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
model.fit(X_train, Y_train)
```

```python
prediction = model.predict(X_test)
print(prediction)
```

```python
coef = model.coef_
print(coef)
```

```python
intercept = model.intercept_
print(intercept)
```

```python
score = model.score(X_test, Y_test)
print(score)
```
