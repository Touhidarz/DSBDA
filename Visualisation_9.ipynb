```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Titanic-Dataset.csv')
```

```python
data['Age'].fillna(data['Age'].mean(), inplace=True)
data['Age'] = data['Age'].astype('int')
```

```python
sns.boxplot(data=data, x='Survived', y='Age', hue='Sex')
```

```python
sns.boxplot(data=data, x='Sex', y='Age', hue='Survived')
```

```python
sns.boxplot(data=data, y='Age')
```

```python
mean = data['Age'].mean()
std = data['Age'].std()

threshold = 3
zScore = (data['Age'] - mean) / std

data['Age'] = data['Age'].mask ( (zScore > threshold) | (zScore<-threshold) ,int(data['Age'].mean()) )
```

```python
sns.boxplot(data=data, y='Age')
```
